```python
# ===============================================================
# Cell 1: 環境設定與套件安裝
# ===============================================================
import sys
import subprocess

packages = ['requests', 'beautifulsoup4', 'pandas', 'lxml']
for package in packages:
    print(f"--- 正在安裝 {package} ---")
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"✅ {package} 安裝完成！\n")
    except subprocess.CalledProcessError as e:
        print(f"❌ {package} 安裝失敗: {e}")

print("🎉 所有必要的套件都已準備就緒！")
```

```python
# ===============================================================
# Cell 2: 全域設定與參數 (邏輯修正版)
# ===============================================================
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import time
import os
import json

# --- 通用設定 ---
HEADERS = {
    'User-Agent': 'T-Forecast-Project-Bot/1.0 (Academic research for a university project)'
}
DB_FILENAME = "cna_articles_db.csv"
HEADLINE_LOG_FILENAME = f"cna_headline_log_{datetime.now().strftime('%Y-%m-%d')}.csv"

# --- 統一的目標 URL ---
CNA_HOME_URL = "https://www.cna.com.tw/"

# --- 爬蟲選擇器 ---
# 內文爬蟲：抓取首頁上「所有」看起來像新聞的連結
ALL_NEWS_SELECTOR_ON_HOME = 'a[href*="/news/"]' 
# 內文爬蟲：抓取文章內頁的 JSON-LD
JSON_LD_SELECTOR = 'script[type="application/ld+json"]'
# 頭條標籤爬蟲：分別抓取 Level 1 和 Level 2 的頭條版位
HEADLINE_LEVEL_1_SELECTOR = "div#biggestNews a.majorNewsClick"
HEADLINE_LEVEL_2_SELECTOR = "div#otherNews a.majorNewsClick"

print("所有參數已更新並載入記憶體 (邏輯修正版)。")
```

```python
# ===============================================================
# Cell 3: 核心爬蟲函式與主要執行流程 (v2 - 新增升級紀錄功能)
# ===============================================================

# --- 函式定義區 (此區塊維持不變) ---

def get_article_urls_from_homepage(homepage_url, selector, existing_urls_set):
    """從首頁抓取所有新聞文章的 URL。"""
    print("正在從首頁抓取所有可見的新文章 URL...")
    new_urls = []
    try:
        response = requests.get(homepage_url, headers=HEADERS, timeout=15)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        
        links = soup.select(selector)
        if not links:
            print("在首頁找不到任何文章連結，請檢查 ALL_NEWS_SELECTOR_ON_HOME。")
            return []

        for link in links:
            url = link.get('href')
            if url and not url.startswith('http'):
                url = 'https://www.cna.com.tw' + url
            
            if url and url not in existing_urls_set and url not in new_urls:
                new_urls.append(url)
        
        print(f"從首頁共發現 {len(new_urls)} 篇新文章。")
        return new_urls
    except Exception as e:
        print(f"抓取首頁 {homepage_url} 失敗: {e}")
        return []

def scrape_article_content_from_jsonld(article_url):
    """訪問單一文章頁面，並從 JSON-LD 中提取結構化資料。"""
    try:
        time.sleep(1)
        response = requests.get(article_url, headers=HEADERS, timeout=15)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        script_tag = soup.find('script', type='application/ld+json')
        if not script_tag: return None
        json_data_list = json.loads(script_tag.string)
        news_data = next((item for item in json_data_list if item.get("@type") == "NewsArticle"), None)
        if not news_data: return None
        return {
            'url': news_data.get('url'), 'title': news_data.get('headline'),
            'content': news_data.get('articleBody'), 'source': news_data.get('publisher', {}).get('name'),
            'published_at': news_data.get('datePublished'), 'scraped_at': datetime.now(),
            'credibility_label': 0, 'headline_level': 0
        }
    except Exception as e:
        print(f"抓取文章 {article_url} 失敗: {e}")
        return None

def scrape_cna_headlines_tags():
    """訪問中央社首頁，抓取頭條版位的 URL 並標記層級。"""
    try:
        time.sleep(2)
        response = requests.get(CNA_HOME_URL, headers=HEADERS, timeout=15)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        results = []
        crawl_time = datetime.now()
        for level, selector in [(1, HEADLINE_LEVEL_1_SELECTOR), (2, HEADLINE_LEVEL_2_SELECTOR)]:
            for link in soup.select(selector):
                url = link.get('href')
                if url and not url.startswith('http'):
                    url = 'https://www.cna.com.tw' + url
                if url:
                    results.append({ 'observed_at': crawl_time, 'url': url, 'headline_level': level })
        return results
    except Exception as e:
        print(f"[{datetime.now()}] 抓取頭條標籤時發生錯誤: {e}")
        return []

# --- 主要執行流程 (包裹在 __main__ 區塊中) ---
def main_execution():
    # Part 1: 抓取新文章內文 (邏輯不變)
    print("--- Part 1: 開始執行內文爬取任務 ---")
    existing_urls = set()
    if os.path.exists(DB_FILENAME):
        db_df_part1 = pd.read_csv(DB_FILENAME)
        existing_urls = set(db_df_part1['url'])
    print(f"主資料庫中已存在 {len(existing_urls)} 篇文章。")
    new_urls_to_scrape = get_article_urls_from_homepage(CNA_HOME_URL, ALL_NEWS_SELECTOR_ON_HOME, existing_urls)
    new_articles_data = []
    if new_urls_to_scrape:
        for url in new_urls_to_scrape:
            print(f"正在抓取內文: {url[:70]}...")
            content = scrape_article_content_from_jsonld(url)
            if content:
                new_articles_data.append(content)
    if new_articles_data:
        new_df = pd.DataFrame(new_articles_data)
        new_df.to_csv(DB_FILENAME, mode='a', header=not os.path.exists(DB_FILENAME), index=False, encoding='utf-8-sig')
        print(f"\n✅ Part 1 任務完成！成功新增 {len(new_articles_data)} 篇文章至 {DB_FILENAME}\n")
    else:
        print("\n🟡 Part 1 任務完成，但沒有發現或成功抓取任何新文章。\n")

    # Part 2: 抓取今日頭條標籤 (邏輯不變)
    print("--- Part 2: 開始執行頭條標籤爬取任務 ---")
    headline_log_data = scrape_cna_headlines_tags()
    if headline_log_data:
        df_log = pd.DataFrame(headline_log_data)
        df_log.to_csv(HEADLINE_LOG_FILENAME, mode='a', header=not os.path.exists(HEADLINE_LOG_FILENAME), index=False, encoding='utf-8-sig')
        print(f"✅ Part 2 任務完成！成功記錄 {len(headline_log_data)} 筆頭條標籤至 {HEADLINE_LOG_FILENAME}\n")
    else:
        print("🟡 Part 2 任務完成，但本次沒有抓取到任何頭條標籤。\n")

    # Part 3: 整合標籤至主資料庫 (⭐ 已加入「升級紀錄」新功能 ⭐)
    print("--- Part 3: 開始執行數據整合與標籤更新任務 ---")
    if os.path.exists(DB_FILENAME):
        db_df_part3 = pd.read_csv(DB_FILENAME)
        update_count = 0
        promoted_articles = [] # <-- 新增：建立一個空列表來記錄升級的文章

        if os.path.exists(HEADLINE_LOG_FILENAME):
            log_df = pd.read_csv(HEADLINE_LOG_FILENAME)
            print(f"開始更新，主資料庫目前有 {db_df_part3.shape[0]} 筆資料...")
            
            # 使用 URL 作為索引，可以大幅加速更新過程
            db_df_part3.set_index('url', inplace=True)
            
            for index, row in log_df.iterrows():
                url = row['url']
                level = int(row['headline_level']) # 確保層級是整數
                
                # 檢查這個 URL 是否存在於主資料庫的索引中
                if url in db_df_part3.index:
                    old_level = int(db_df_part3.loc[url, 'headline_level']) # <-- 新增：取得舊的層級

                    # <-- 新增：檢查是否從 Level 2 升級到 Level 1
                    if old_level == 2 and level == 1:
                        promoted_info = {
                            'promotion_time': row['observed_at'],
                            'url': url,
                            'title': db_df_part3.loc[url, 'title'] 
                        }
                        promoted_articles.append(promoted_info)
                        # 在終端機即時顯示升級訊息
                        print(f"📈 頭條升級: {promoted_info['title'][:30]}...")

                    # 只有當新的層級更高時，才進行更新
                    if old_level < level:
                        db_df_part3.loc[url, 'headline_level'] = level
                        update_count += 1
            
            db_df_part3.reset_index(inplace=True)
        
        # 將更新後的 DataFrame 存回 CSV 檔案
        db_df_part3.to_csv(DB_FILENAME, index=False, encoding='utf-8-sig')
        
        # <-- 新增：在迴圈結束後，檢查是否有升級紀錄，並存檔
        if promoted_articles:
            promo_df = pd.DataFrame(promoted_articles)
            promo_filename = f"cna_promoted_headlines_{datetime.now().strftime('%Y-%m-%d')}.csv"
            promo_df.to_csv(promo_filename, mode='a', header=not os.path.exists(promo_filename), index=False, encoding='utf-8-sig')
            print(f"🏅 已將 {len(promoted_articles)} 筆升級紀錄存至 {promo_filename}")

        print(f"✅ Part 3 整合完畢！共更新了 {update_count} 筆資料的頭條標籤。")
    else:
        print("🟡 Part 3 整合失敗，找不到主資料庫檔案。請先確保 Part 1 成功執行。")

# 讓 Cell 可以被直接執行
if __name__ == "__main__" or "ipykernel" in sys.modules:
    main_execution()
```

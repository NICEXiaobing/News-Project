```python
import pandas as pd
import numpy as np
import time
import platform
import matplotlib.pyplot as plt
import seaborn as sns

# Scikit-learn
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
from scipy.sparse import hstack

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
import lightgbm as lgb

# 設定
INPUT_CSV = 'News_Data_tokenized.csv' 
TFIDF_MAX_FEATURES = 50000
SEED = 42
CV_FOLDS = 10  # 10-Fold
NUM_CORES = 12

# 字型設定
if platform.system() == 'Darwin':
    plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'Heiti TC']
else:
    plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False

if __name__ == '__main__':
    print(f"🚀 [Step 2] 啟動 7 大模型訓練 (80-20 Split -> 10-Fold on Train)...")
    
    # 1. 讀取資料
    try: df = pd.read_csv(INPUT_CSV)
    except: 
        print(f"❌ 找不到 {INPUT_CSV}，請先執行程式一！")
        exit()
        
    df = df.dropna(subset=['tokens_str', 'label'])
    
    # 若 CSV 缺特徵欄位則補算 (防呆)
    if 'title_len' not in df.columns:
        df['title_len'] = df['title_clean'].astype(str).apply(len)
        df['punct_count'] = df['title_clean'].astype(str).apply(lambda x: x.count('!')+x.count('?'))
        df['has_digit'] = df['title_clean'].astype(str).apply(lambda x: 1 if any(c.isdigit() for c in x) else 0)
    
    df[['title_len', 'punct_count', 'has_digit']] = df[['title_len', 'punct_count', 'has_digit']].fillna(0)
    print(f"   有效資料: {len(df)} 筆")

    # 2. 特徵工程 (Feature Engineering)
    # A. TF-IDF
    tfidf = TfidfVectorizer(max_features=TFIDF_MAX_FEATURES, token_pattern=r'(?u)\b\w+\b')
    X_text = tfidf.fit_transform(df['tokens_str'].astype(str))
    
    # B. Numeric
    scaler = MinMaxScaler()
    X_stats = scaler.fit_transform(df[['title_len', 'punct_count', 'has_digit']])
    
    # C. Hybrid
    X_hybrid = hstack([X_text, X_stats])
    X_text_only = X_text # 給 Naive Bayes
    y = df['label'].values

    # 3. 資料切分 (先切出 80% 訓練集, 20% 測試集)
    print(f"   執行資料切分 (Test Size: 20%)...")
    X_train, X_test, y_train, y_test = train_test_split(X_hybrid, y, test_size=0.2, random_state=SEED, stratify=y)
    X_train_nb, X_test_nb, _, _ = train_test_split(X_text_only, y, test_size=0.2, random_state=SEED, stratify=y)
    
    print(f"   Train Set: {X_train.shape[0]} | Test Set: {X_test.shape[0]}")

    # 4. 定義模型
    models = {
        "Logistic Regression": (LogisticRegression(max_iter=1000, class_weight='balanced'), False),
        "Naive Bayes": (MultinomialNB(), True), 
        "Linear SVM": (LinearSVC(class_weight='balanced', dual='auto', max_iter=2000), False),
        "Decision Tree": (DecisionTreeClassifier(class_weight='balanced', max_depth=20), False),
        "Random Forest": (RandomForestClassifier(n_estimators=100, class_weight='balanced', n_jobs=NUM_CORES), False),
        "XGBoost": (xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', scale_pos_weight=10, n_jobs=NUM_CORES), False),
        "LightGBM": (lgb.LGBMClassifier(class_weight='balanced', n_jobs=NUM_CORES, verbose=-1), False)
    }

    # 儲存所有結果的容器
    all_results = []

    print(f"\n🥊 開始訓練 (計算 10-Fold CV 與 測試集指標)...")
    
    for name, (model, use_text_only) in models.items():
        print(f"   Running {name}...", end="\r") # 動態顯示進度
        
        # 選擇特徵集
        X_tr = X_train_nb if use_text_only else X_train
        X_te = X_test_nb if use_text_only else X_test
        
        # A. 10-Fold Cross Validation (只在 80% Train Set 上做)
        cv_scores = cross_val_score(model, X_tr, y_train, cv=CV_FOLDS, scoring='f1')
        cv_f1_mean = cv_scores.mean()
        
        # B. 完整訓練與預測 (Train on 80%, Test on 20%)
        model.fit(X_tr, y_train)
        y_pred = model.predict(X_te)
        
        # C. 計算機率 (為了 AUROC)
        try:
            if hasattr(model, "predict_proba"):
                y_prob = model.predict_proba(X_te)[:, 1]
            elif hasattr(model, "decision_function"):
                y_prob = model.decision_function(X_te)
            else:
                y_prob = y_pred 
            auroc = roc_auc_score(y_test, y_prob)
        except:
            auroc = 0.5

        # D. 計算 5 大指標
        acc = accuracy_score(y_test, y_pred)
        # binary average 用於 0/1 分類
        prec = precision_score(y_test, y_pred, average='binary') 
        rec = recall_score(y_test, y_pred, average='binary')
        f1 = f1_score(y_test, y_pred, average='binary')
        
        # E. 生成詳細分類報告 (包含 Macro, Weighted Avg)
        report_dict = classification_report(y_test, y_pred, output_dict=True)
        report_str = classification_report(y_test, y_pred, digits=4) # 轉成文字供列印
        
        # F. 混淆矩陣
        cm = confusion_matrix(y_test, y_pred)
        
        # 存入列表
        all_results.append({
            "Model": name,
            "Accuracy": acc,
            "Precision": prec,
            "Recall": rec,
            "F1-Score": f1,
            "AUROC": auroc,
            "CV_F1": cv_f1_mean,
            "Report_Str": report_str,
            "CM": cm
        })

    print(f"   訓練完成！準備輸出報告。{' '*20}")
    
    # ==========================================
    # 輸出部分 1: 所有模型訓練結果排行 (5大指標)
    # ==========================================
    # 轉為 DataFrame 並依照 F1-Score 排序
    df_rank = pd.DataFrame(all_results)[['Model', 'Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUROC']]
    df_rank = df_rank.sort_values(by='F1-Score', ascending=False)
    
    print("\n" + "="*80)
    print("🏆 【 所有模型訓練結果排行：五大指標成績表 】")
    print("="*80)
    # 使用 to_string 確保對齊與小數點格式
    print(df_rank.to_string(index=False, float_format="{:.4f}".format, col_space=12))
    print("\n")

    # ==========================================
    # 輸出部分 2: 各個模型的詳細結果
    # ==========================================
    # 依照上面的排名順序依序輸出
    for idx, row in df_rank.iterrows():
        model_name = row['Model']
        # 找回原本存的詳細資料
        original_data = next(item for item in all_results if item["Model"] == model_name)
        
        print("="*80)
        print(f"📌 模型：{model_name}")
        print("-" * 80)
        
        # 1. 印出分類報告 (Precision, Recall, F1, Support)
        print("【 三大指標詳細成績表 】")
        print(original_data['Report_Str'])
        print("-" * 80)
        
        # 2. 繪製混淆矩陣圖
        print(f"【 混淆矩陣圖 - {model_name} 】")
        cm = original_data['CM']
        
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={"size": 14})
        plt.title(f'{model_name}\nF1: {row["F1-Score"]:.3f}', fontsize=14)
        plt.ylabel('Actual Label', fontsize=12)
        plt.xlabel('Predicted Label', fontsize=12)
        plt.xticks([0.5, 1.5], ['L0', 'L1'])
        plt.yticks([0.5, 1.5], ['L0', 'L1'])
        plt.tight_layout()
        plt.show() # 直接顯示，不存檔
        print("\n\n")

    print("✅ 所有結果輸出完畢 (無產生外部檔案)")
```

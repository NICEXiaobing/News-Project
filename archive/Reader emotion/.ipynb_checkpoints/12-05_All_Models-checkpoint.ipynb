```python
# step2_train_evaluate.py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import gc
import warnings
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder, LabelBinarizer
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score, f1_score, precision_score, recall_score

# 模型庫
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.calibration import CalibratedClassifierCV

warnings.filterwarnings('ignore')

# ==========================================
# 訓練與評估函式 (不畫圖，只回傳數據)
# ==========================================
def train_and_evaluate(name, model, X_train, y_train, X_test, y_test, class_names):
    print(f"   -> 正在訓練: {name} ...", end="\r")
    start_t = time.time()
    
    try:
        # 訓練
        model.fit(X_train, y_train)
        train_time = time.time() - start_t
        
        # 預測
        y_pred = model.predict(X_test)
        
        # 嘗試取得機率 (AUROC用)
        y_pred_proba = None
        if hasattr(model, "predict_proba"):
            try: y_pred_proba = model.predict_proba(X_test)
            except: pass
            
        # 計算指標 (Weighted Average)
        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        rec = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
        
        auroc = 0.0
        if y_pred_proba is not None:
            try:
                lb = LabelBinarizer()
                lb.fit(y_train)
                y_test_bin = lb.transform(y_test)
                auroc = roc_auc_score(y_test_bin, y_pred_proba, multi_class='ovr', average='macro')
            except: pass
        
        # 混淆矩陣
        cm = confusion_matrix(y_test, y_pred)
        
        print(f"   ✅ {name} 完成 (耗時: {train_time:.1f}s) | Acc: {acc:.4f}")
        
        return {
            "Model": name,
            "Accuracy": acc,
            "Precision": prec,
            "Recall": rec,
            "F1_Score": f1,
            "AUROC": auroc,
            "Time": train_time,
            "Confusion_Matrix": cm,
            "Report": classification_report(y_test, y_pred, target_names=class_names, zero_division=0)
        }

    except Exception as e:
        print(f"   ❌ {name} 失敗: {e}")
        return None

# ==========================================
# 主程式
# ==========================================
if __name__ == '__main__':
    main_start = time.time()
    print("🚀 [步驟二] 七大模型全速評測 (Train/Test 分離版)")
    print("   - 硬體配置: M4 Pro (全核全速)")
    print("   - 特徵工程: TF-IDF 20,000 維度 (高精度)")
    print("   - KNN 設定: 不閹割 (n_jobs=-1)")

    # 1. 讀取已處理的資料
    try:
        df_train = pd.read_csv('Processed_Train.csv')
        df_test = pd.read_csv('Processed_Test.csv')
        
        df_train['text_segmented'] = df_train['text_segmented'].fillna('')
        df_test['text_segmented'] = df_test['text_segmented'].fillna('')
        
        print(f"✅ 資料載入: 訓練集 {len(df_train)} 筆 / 測試集 {len(df_test)} 筆")
    except:
        print("❌ 找不到 Processed_*.csv，請先執行步驟一")
        exit()

    # 2. 標籤編碼 (Label Encoding)
    le = LabelEncoder()
    # 必須用訓練集的標籤來 fit
    y_train = le.fit_transform(df_train['Column1'])
    # 測試集如果有沒看過的標籤會報錯，這裡假設測試集標籤都在訓練集中
    # 實務上應該處理 unknown，但這裡先假設資料完整
    y_test = le.transform(df_test['Column1'])
    class_names = le.classes_

    # 3. 特徵工程 (TF-IDF)
    # 根據 M4 Pro 效能，我們開到 20,000 維度
    print("🔄 TF-IDF 轉換 (Max 20,000, float32)...")
    vectorizer = TfidfVectorizer(
        max_features=20000, 
        ngram_range=(1, 2), 
        min_df=2, 
        sublinear_tf=True,
        dtype=np.float32
    )
    
    # 嚴格遵守：只在 Train 上 fit，在 Test 上 transform
    X_train = vectorizer.fit_transform(df_train['text_segmented'])
    X_test = vectorizer.transform(df_test['text_segmented'])
    
    # 排序索引 (優化 SVM/Linear 效能)
    X_train.sort_indices()
    X_test.sort_indices()
    
    del df_train, df_test
    gc.collect()

    # 4. 定義模型 (全速模式)
    models = [
        ("Naive Bayes", MultinomialNB()),
        
        ("Logistic Regression", LogisticRegression(max_iter=1000, class_weight='balanced', n_jobs=-1, random_state=42)),
        
        ("Decision Tree", DecisionTreeClassifier(max_depth=50, class_weight='balanced', random_state=42)),
        
        ("SVM (Linear)", CalibratedClassifierCV(LinearSVC(dual=False, class_weight='balanced', random_state=42), n_jobs=-1)),
        
        # [KNN 全速] 注意：這會非常吃記憶體
        ("KNN", KNeighborsClassifier(n_neighbors=5, n_jobs=-1)), 
        
        ("XGBoost", XGBClassifier(n_estimators=200, eval_metric='mlogloss', n_jobs=-1, tree_method='hist', random_state=42)),
        
        ("LightGBM", LGBMClassifier(n_estimators=200, class_weight='balanced', n_jobs=-1, verbose=-1, random_state=42))
    ]

    # 5. 執行訓練迴圈
    results_data = []
    print("\n🏋️ 開始依序訓練模型...")
    
    for name, model in models:
        res = train_and_evaluate(name, model, X_train, y_train, X_test, y_test, class_names)
        if res: results_data.append(res)
        gc.collect()

    # 6. 輸出結果
    if results_data:
        # --- A. 排行榜 ---
        df_results = pd.DataFrame(results_data)
        df_results = df_results[['Model', 'Accuracy', 'F1_Score', 'Precision', 'Recall', 'AUROC', 'Time']]
        df_results = df_results.sort_values(by='Accuracy', ascending=False).reset_index(drop=True)
        
        print("\n" + "="*80)
        print("🏆 模型效能排行榜 (由高到低)")
        print("="*80)
        print(df_results.to_string(index=False, float_format="%.4f"))
        print("="*80)
        
        # --- B. 混淆矩陣圖 ---
        print("\n📊 依序輸出混淆矩陣...")
        # 依照排名順序輸出圖表
        for i, row in df_results.iterrows():
            name = row['Model']
            # 找出對應的原始資料
            res = next(item for item in results_data if item["Model"] == name)
            
            print(f"\n🔹 第 {i+1} 名: {name}")
            print(res['Report']) # 印出詳細報告
            
            plt.figure(figsize=(8, 6))
            sns.heatmap(res['Confusion_Matrix'], annot=True, fmt='d', cmap='RdPu', 
                        xticklabels=class_names, yticklabels=class_names)
            plt.title(f'{name} Confusion Matrix (Acc: {row["Accuracy"]:.4f})')
            plt.ylabel('True Label')
            plt.xlabel('Predicted Label')
            plt.show()
    
    print(f"\n✅ 全部流程結束！總耗時: {(time.time() - main_start)/60:.1f} 分鐘")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import pickle

# Sklearn 相關
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder

# 模型
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# 設定中文字型 (避免混淆矩陣亂碼)
import matplotlib.font_manager as fm
# 嘗試設定常見的繁體中文字體，若無效請自行更換路徑
plt.rcParams['font.sans-serif'] = ['Taipei Sans TC Beta', 'Microsoft JhengHei', 'SimHei', 'Arial Unicode MS'] 
plt.rcParams['axes.unicode_minus'] = False

# ==========================================
# 設定區
# ==========================================
TRAIN_PKL = 'train_seg.pkl'
TEST_PKL = 'test_seg.pkl'
NUM_CORES = 12

# ==========================================
# 主程式
# ==========================================
if __name__ == '__main__':
    print("1. 讀取處理後的資料...")
    try:
        train_df = pd.read_pickle(TRAIN_PKL)
        test_df = pd.read_pickle(TEST_PKL)
    except FileNotFoundError:
        print("錯誤：找不到 pkl 檔案，請先執行前處理程式。")
        exit()

    X_train_raw = train_df['text'].fillna('')
    y_train_raw = train_df['label']
    X_test_raw = test_df['text'].fillna('')
    y_test_raw = test_df['label']

    print(f"   訓練集數量: {len(X_train_raw)}, 測試集數量: {len(X_test_raw)}")

    # 標籤編碼 (Label Encoding)
    le = LabelEncoder()
    y_train = le.fit_transform(y_train_raw)
    y_test = le.transform(y_test_raw)
    label_names = le.classes_
    print(f"   類別標籤: {label_names}")

    # 特徵工程: TF-IDF
    print("2. 執行 TF-IDF 特徵提取...")
    # max_features 設定為 5000 以節省記憶體並加速，若記憶體足夠可移除此參數或設更大
    tfidf = TfidfVectorizer(max_features=10000, token_pattern=r'(?u)\b\w+\b') 
    X_train = tfidf.fit_transform(X_train_raw)
    X_test = tfidf.transform(X_test_raw)
    print(f"   TF-IDF 矩陣形狀: {X_train.shape}")

    # 定義模型
    # 注意: SVM 使用 LinearSVC 因為傳統 SVC(kernel='rbf') 在大量文本特徵下會非常慢且吃記憶體
    models = {
        'NB': MultinomialNB(),
        'LR': LogisticRegression(n_jobs=NUM_CORES, max_iter=1000),
        'DT': DecisionTreeClassifier(),
        'SVM': LinearSVC(max_iter=2000, dual='auto'), 
        'KNN': KNeighborsClassifier(n_jobs=NUM_CORES),
        'XGB': XGBClassifier(n_jobs=NUM_CORES, eval_metric='mlogloss'),
        'LGBM': LGBMClassifier(n_jobs=NUM_CORES, verbose=-1)
    }

    # 儲存結果
    results = []
    
    # 10-Fold Cross Validation 設定
    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

    print("\n3. 開始 10-Fold 訓練與評估 (依序執行)...")
    
    for name, model in models.items():
        print(f"   正在處理模型: {name} ...")
        start_time = time.time()
        
        # 3.1: 10-Fold Cross Validation (訓練集內部評估)
        # n_jobs=NUM_CORES 讓交叉驗證平行運算
        cv_scores = cross_val_score(model, X_train, y_train, cv=skf, scoring='accuracy', n_jobs=NUM_CORES)
        avg_acc = np.mean(cv_scores)
        
        # 3.2: 全量訓練並預測測試集 (產生混淆矩陣用)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        test_acc = accuracy_score(y_test, y_pred)
        
        elapsed_time = time.time() - start_time
        print(f"      -> CV平均準確率: {avg_acc:.4f} | 測試集準確率: {test_acc:.4f} | 耗時: {elapsed_time:.2f}秒")
        
        results.append({
            'Model': name,
            'CV_Accuracy': avg_acc,
            'Test_Accuracy': test_acc,
            'Time_Sec': elapsed_time
        })

        # 3.3: 繪製混淆矩陣
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_names, yticklabels=label_names)
        plt.title(f'{name} Confusion Matrix (Test Set)\nAcc: {test_acc:.4f}')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.savefig(f'cm_{name}.png') # 儲存圖片
        plt.close()
        
        # 輸出單一模型的混淆矩陣表格 (CSV)
        cm_df = pd.DataFrame(cm, index=[f"True_{x}" for x in label_names], columns=[f"Pred_{x}" for x in label_names])
        cm_df.to_csv(f'cm_table_{name}.csv', encoding='utf-8-sig')

    # 4. 輸出總排行表
    print("\n4. 產生總排行表...")
    results_df = pd.DataFrame(results)
    results_df = results_df.sort_values(by='CV_Accuracy', ascending=False)
    
    print("\n========== 模型訓練結果總排行 ==========")
    print(results_df)
    results_df.to_csv('model_ranking_summary.csv', index=False, encoding='utf-8-sig')
    
    print("\n全部任務完成！請查看生成的 .csv 表格與 .png 圖片。")
```

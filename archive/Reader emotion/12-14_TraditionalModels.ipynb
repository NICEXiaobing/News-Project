```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time

# Sklearn 相關
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    accuracy_score, confusion_matrix, classification_report,
    precision_recall_fscore_support, roc_auc_score
)
from sklearn.preprocessing import LabelEncoder
from sklearn.calibration import CalibratedClassifierCV

# 模型
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# 設定中文字型 (確保圖表顯示正常)
import matplotlib.font_manager as fm
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'Microsoft JhengHei', 'SimHei', 'sans-serif'] 
plt.rcParams['axes.unicode_minus'] = False

# ==========================================
# 設定區
# ==========================================
TRAIN_PKL = 'train_seg.pkl'
TEST_PKL = 'test_seg.pkl'
NUM_CORES = 12

# ==========================================
# 主程式
# ==========================================
if __name__ == '__main__':
    print("1. 讀取資料中...")
    try:
        train_df = pd.read_pickle(TRAIN_PKL)
        test_df = pd.read_pickle(TEST_PKL)
    except FileNotFoundError:
        print("錯誤：找不到 pkl 檔案，請先執行前處理程式 (01_preprocessing.py)。")
        exit()

    X_train_raw = train_df['text'].fillna('')
    y_train_raw = train_df['label']
    X_test_raw = test_df['text'].fillna('')
    y_test_raw = test_df['label']

    # Label Encoding
    le = LabelEncoder()
    y_train = le.fit_transform(y_train_raw)
    y_test = le.transform(y_test_raw)
    label_names = le.classes_
    print(f"   類別標籤 ({len(label_names)}類): {label_names}")

    # TF-IDF 特徵工程
    print("2. 執行 TF-IDF 特徵提取...")
    tfidf = TfidfVectorizer(max_features=30000, token_pattern=r'(?u)\b\w+\b') 
    X_train = tfidf.fit_transform(X_train_raw)
    X_test = tfidf.transform(X_test_raw)

    # 定義模型
    # SVM 使用 CalibratedClassifierCV 以支援 predict_proba (算 AUROC 用)
    models = {
        'NB': MultinomialNB(),
        'LR': LogisticRegression(n_jobs=NUM_CORES, max_iter=1000),
        'DT': DecisionTreeClassifier(),
        'SVM': CalibratedClassifierCV(LinearSVC(max_iter=2000, dual='auto')), 
        'KNN': KNeighborsClassifier(n_jobs=NUM_CORES),
        'XGB': XGBClassifier(n_jobs=NUM_CORES, eval_metric='mlogloss'),
        'LGBM': LGBMClassifier(n_jobs=NUM_CORES, verbose=-1)
    }

    # 用於暫存所有模型的結果，以便最後統一輸出
    all_results = []

    print("\n3. 開始訓練與評估 (12核全速運行)...")
    
    for name, model in models.items():
        print(f"   正在訓練模型: {name} ...")
        start_time = time.time()
        
        # 全量訓練
        model.fit(X_train, y_train)
        train_time = time.time() - start_time
        
        # 預測
        y_pred = model.predict(X_test)
        
        # 預測機率 (處理 AUROC 用)
        try:
            y_proba = model.predict_proba(X_test)
            # 計算 Weighted AUROC (One-vs-Rest)
            auroc = roc_auc_score(y_test, y_proba, multi_class='ovr', average='weighted')
        except:
            auroc = 0.0 # 若無法計算則填 0
        
        # 計算整體指標 (Weighted Average)
        acc = accuracy_score(y_test, y_pred)
        prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted', zero_division=0)
        
        # 計算詳細 Class-wise 報表
        class_report = classification_report(y_test, y_pred, target_names=label_names, output_dict=True, zero_division=0)
        class_report_df = pd.DataFrame(class_report).transpose()
        
        # 暫存結果
        all_results.append({
            'Model': name,
            'Summary': {
                'Accuracy': acc,
                'Precision': prec,
                'Recall': rec,
                'F1-score': f1,
                'AUROC': auroc,
                'Time': train_time
            },
            'Detailed_Table': class_report_df,
            'Confusion_Matrix': confusion_matrix(y_test, y_pred),
            'Predictions': y_pred
        })

    # ==========================================
    # 4. 輸出結果 (順序：排行榜 -> 各模型詳情)
    # ==========================================
    
    # 4.1 總排行榜
    print("\n" + "="*60)
    print("🏆 模型訓練結果總排行 (Model Ranking Summary)")
    print("="*60)
    
    ranking_data = []
    for res in all_results:
        s = res['Summary']
        ranking_data.append({
            'Model': res['Model'],
            'Accuracy': s['Accuracy'],
            'Precision': s['Precision'],
            'Recall': s['Recall'],
            'F1-score': s['F1-score'],
            'AUROC': s['AUROC'],
            'Time(s)': s['Time']
        })
    
    ranking_df = pd.DataFrame(ranking_data)
    # 依照 Accuracy 排序
    ranking_df = ranking_df.sort_values(by='Accuracy', ascending=False).reset_index(drop=True)
    print(ranking_df.to_string(float_format="{:.4f}".format))
    print("\n")

    # 4.2 各模型詳細報表與圖表
    # 依照排行榜順序輸出
    for idx, row in ranking_df.iterrows():
        model_name = row['Model']
        # 找到對應的結果資料
        res = next(item for item in all_results if item['Model'] == model_name)
        
        print("="*60)
        print(f"📊 模型詳情: {model_name}")
        print("="*60)
        
        # 顯示詳細比較表
        print(f"訓練時間: {res['Summary']['Time']:.2f} 秒")
        print(f"整體 AUROC: {res['Summary']['AUROC']:.4f}")
        print("\n[ 詳細分類指標表 ]")
        # 調整顯示格式
        print(res['Detailed_Table'].to_string(float_format="{:.4f}".format))
        
        # 顯示混淆矩陣圖
        print(f"\n[ 混淆矩陣圖 - {model_name} ]")
        cm = res['Confusion_Matrix']
        acc = res['Summary']['Accuracy']
        f1 = res['Summary']['F1-score']
        
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_names, yticklabels=label_names)
        plt.title(f'{model_name} Confusion Matrix\nAcc: {acc:.4f} | F1: {f1:.4f}')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.show() # 直接顯示圖片
        print("\n" + "-"*60 + "\n")
```
